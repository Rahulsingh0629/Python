# Object And Class

```python
class Student:              # Class Definition
    def __init__(self):     # Constructor
        self.name = "John Doe"
        self.gender = "Male"
        self.age = 20
    
    def talk(self):                          # Instance Method
        print("Hello, my name is", self.name)

    def vote(self):
        if self.age < 18:
            print("I am not eligible to vote.")
        else:
            print("I am eligible to vote.")
obj = Student()                                  # Object Creation
# obj.talk()
# obj.vote()
Student.talk(obj)                             # Calling Instance Method using Object
Student.vote(obj)
        
```

```python
class Student:
    def __init__(self, n, g, a):
        self.name = n
        self.gender = g
        self.age = a

    def talk(self):
        print("Hello, my name is", self.name)

    def vote(self):
        if self.age < 18:
            print("I am not eligible to vote.")
        else:
            print("I am eligible to vote.")

obj1 = Student("Alice", "Female", 25)
obj2 = Student("Bob", "Male", 18)
obj1.talk()
obj1.vote()
obj2.talk()
obj2.vote()
```

```python
class Person:
    def __init__ (self):
        self.name = str(input("Enter name: "))
        self.gender = str(input("Enter your gender: "))
        self.age = int(input("Enter your age: "))

    def display(self):
        print("Name", self.name)
        print("Gender", self.gender)
        print("Age", self.age)

    def talk(self):
        print("I am", self.name)
    
    def drive(self):
        if self.age < 18:
            print("I am not eligible to drive.")
        else:
            print("I am eligible to drive.")

obj = Person()
Person.display(obj)
obj.talk()
obj.drive()
```

# Inheritance

```python
class Animal:                    # Parent class
    def __init__(self, name):    # Constructor
        self.name = name
 
    def speak(self):                 # Creating a method
        return f"{self.name} make soounds."
    
class Dog(Animal):                # Create Child Class and inherit Parent class (Means Call the Parent class)
    def speak(self):                # Overriding the speak method of Parent class
        return f"{self.name} says Woof!"
    
class Cat(Animal):               # Create Child Class and inherit Parent class (Means Call the Parent class)
    def speak(self):             # Overriding the speak method of Parent class
        return f"{self.name} says Meow!"
    

    
obj = Animal("Dog")   # It can call the speak method of Animal class (Parent class)
dog = Dog("Buddy")    # It can call the speak method of Dog class (Child class)
cat = Cat("Whiskers")  # It can call the speak method of Cat class (Child class)
print(obj.speak())     # It calls the speak method of Animal class
print(dog.speak())
print(cat.speak())
```
