# For Loop

```python
sum = 0
for i in range(0, 10):
    if i % 2 == 0:
        sum = sum + i
print("Sum of even numbers from 0 to 50 is:", sum)
```

```python
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
r = int(input("Enter the number of rows: "))
c = int(input("Enter the number of columns: "))
x = []
val = []
for i in range(0, r):
    for j in range(0, c):
        val.insert(j, int(input("Enter the %d * %d element: " % (i, j))))
    x.insert(i, val)
    val = []

y = []
for i in range(0, r):
    for j in range(0, c):
        val.insert(j, x[j][i])
    y.insert(i, val)
    val = []

sum = []
for i in range(0, r):
    for j in range(0, c):
        val.insert(j, x[j][i] + y[i][j])
    sum.insert(i, val)
    val = []

print(sum)
```

# While Loop


```python
n = 1
while n <= 10:
    print("Hello", end=" ")
    n +=1
```

```python
m = 10
while m >= 1:
    print("World", end=" ")

    m -= 1
```

```python
num = int(input("Enter a number: "))
sum = 0
while num <= n:
    sum = sum + num
    num += 1
print("The sum is:", sum)
```

```python
n = int(input("Enter a number: "))
nr = 0
while n%10 !=0:
    c = n%10
    nr = nr *10 + c
    n = n//10
print("The reversed number is:", nr)
```

```python
x = [1,2,3,"Hellow World"]
length = 0
i = 0
try:
    while True:
        # access element by index to trigger IndexError when out of range
        _ = x[i]
        length += 1
        i += 1
except IndexError:
    print(length)
```

```python
import random

nump = random.randint(1000, 9999)
# print(nump) # For testing purposes, to see the generated number
n = int(input("Enter a 4-digit number: "))

while n != 10:
    num = nump
    cor = 0
    while num % 10 :
        numc = num%10
        nc = n % 10
        num = num // 10
        n = n // 10
        if numc == nc:
            cor += 1
    if cor == 4:
        print("Congratulations! You guessed the number.", nump)
        break
    else:
        print("%d digits are correct." % cor)
        n = int(input("Enter a 4-digit number: "))
else:
    print("Game Over! The number was:", nump)
```
