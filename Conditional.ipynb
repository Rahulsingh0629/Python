Nested If

```python
num = 50
if num > 40 == 0:
    if num % 2 == 0:
        print("num is even number but less than 40")
    else:
        print("num is odd number but less than 40")
else:
    print("num is greater than 40")
```

```python
n = 15
if n < 25:
    if n % 2 == 0:
        print("n is even and less than 25")
    else:
        print("n is odd and less than 25")
else:
    print("n is greater than or equal to 25")
```

If-Elif-Else Statement

```python
var = "N"
if var == "A":
    print("This is the vowel A")
elif var == "E":
    print("This is the vowel E")
elif var == "I":
    print("This is the vowel I")
elif var == "O":
    print("This is the vowel O")
elif var == "U":
    print("This is the vowel U")
else:
    print("This is a consonant")
```

# While Loop

```python
count = 1
while count<= 100:
    print(count, end = " ")
    count += 1
```

Break Statement

```python

num = 0
while num < 10:
    num += 1
    if num == 5:
        continue # Skip the printing when num is 5
    if num == 9:
        break # Exit the loop when num is 9
    print(num, end = " ")
```

Using else with While Loop

```python


x = 10
while x > 0:
    print("Coundown:", x)
    x -= 1
else:
    print("Loop end the execution normally.")
```

# Nested for Loop

```python
x = [[1,2,3,4,5],["a","b","c","d","e"],]
for i in x:
    for j in i:
        print(j, end= " ")
    print()
```
