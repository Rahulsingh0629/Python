# string is a sequence of characters

```python
str = 'Tim said, "It\'s a beautiful day!"'
```

```python
print(str)
```

```python
name = '''Hello Programmers,
Welcome to the world of Python programming.
Have a nice day!'''
print(name)
```

```python
print(len(name))
print(name[0:5]) # you can also use only number to print from start or to end like print(name[:5]) or print(name[5:])
print(name.lower())
print(name.upper())
print(name.strip()) # removes any extra spaces from start and end
print(name.replace("Python", "Java"))
print(name.split(",")) # splits the string at the specified separator and returns a list
print("Python" in name) # returns True if "Python" is found in the string
print("Java" in name) # returns True if "Java" is not found in the string
```

```python
for i in name:
    print(i, end="")
```

```python
print(name + str)

stg = "\n{} \n{}".format(name, str)
print(stg)
```

# Dictionaries 

```python
d1 = {
    1 : "Hello",
    2: "Programmers",
    3 : "Welcome to Python",
}

d2 = {
    'Name' : "Alice",
    'age': 30,
    'city': "New York"
}
print(d1)
print(d2)
```

```python
d3 = dict({
    "My name is":"Bob",
    "I am":"a developer",
})
print(d3)
```

```python
d4 = dict([(1, "One"), (2, "Two"), (3, "Three")])
print(d4)

d5 = dict.fromkeys(['a', 'b', 'c'], 0)
print(d5)

d6 = dict({"name":{"first":"John", "last":"Doe"}, "age":25})
print(d6)
```

# Adding Elements

```python
d = {}
d[0] = "Welcome"
print(d)
d[1]= ("to", "Python")
print(d)
d["name"]= {"first":"Jane", "last":"Doe"}
print(d)
```

# Sets 
#It is an unorder collection of unique element

```python
s = set([1, 2, 3, 4, 5])
print(s)
```

```python
s.add("A")
print(s)
```

```python
fs = frozenset([1, 2, 3, 4, 5])
print(fs)
```

```python
fs.add("A")
```

```python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}

print(s1 | s2) # union                        print(s1.union(s2))
print(s1 & s2) # intersection                 print(s1.intersection(s2))
print(s1 - s2) # difference                   print(s1.difference(s2))
print(s1 ^ s2) # symmetric difference         print(s1.symmetric_difference(s2))
```
